```python
import pandas as pd
import json
import pathlib
```

```python
with open("Data 2018-2023/Project/2023/202302889") as file:
    data = json.load(file)
```

```python
print(json.dumps(data, indent=2))
```

```python
# Author
pd.json_normalize(
    data,
    record_path=['abstracts-retrieval-response', 'authors', 'author'],
    errors='ignore',
)
```

```python
# Title
data['abstracts-retrieval-response']['coredata']['dc:title']
```

```python
# Abstract
print(data['abstracts-retrieval-response']['item']['bibrecord']['head']['abstracts'])
```
